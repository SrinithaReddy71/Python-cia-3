```python
import datetime

class Team:

    def __init__(self, team_name, captain_name, num_batsmen, num_bowlers, num_allrounders):
        self.team_name = team_name
        self.captain_name = captain_name
        self.num_batsmen = num_batsmen
        self.num_bowlers = num_bowlers
        self.num_allrounders = num_allrounders

    def display(self):
        print(f"Team: {self.team_name}, Captain: {self.captain_name}, Batsmen: {self.num_batsmen}, Bowlers: {self.num_bowlers}, All-rounders: {self.num_allrounders}")

class MatchRecord(Team):
    
    def __init__(self, team_name, captain_name, num_batsmen, num_bowlers, num_allrounders, opponent_team, runs_scored, match_result, match_date):
        super().__init__(team_name, captain_name, num_batsmen, num_bowlers, num_allrounders)
        self.opponent_team = opponent_team
        self.runs_scored = runs_scored
        self.match_result = match_result
        self.match_date = match_date

    def display(self):
        super().display()
        print(f"Opponent: {self.opponent_team}, Runs Scored: {self.runs_scored}, Result: {self.match_result}, Date: {self.match_date}")

class Player:
   
    def __init__(self, player_id, player_name, team_name):
        self.player_id = player_id
        self.player_name = player_name
        self.team_name = team_name

    def display(self):
        print(f"Player ID: {self.player_id}, Name: {self.player_name}, Team: {self.team_name}")

class TournamentSystem:
    
    def __init__(self):
        self.teams = {}  # Dict of Team objects, key is team_name
        self.players = {} # Dict of Player objects, key is player_id
        self.match_records = [] # List of MatchRecord objects

    def add_team(self):
       
        try:
            team_name = input("Enter team name: ")
            if team_name in self.teams:
                print(f"Error: Team '{team_name}' already exists.")
                return
            captain_name = input("Enter captain's name: ")
            num_batsmen = int(input("Enter number of batsmen: "))
            num_bowlers = int(input("Enter number of bowlers: "))
            num_allrounders = int(input("Enter number of all-rounders: "))

            new_team = Team(team_name, captain_name, num_batsmen, num_bowlers, num_allrounders)
            self.teams[team_name] = new_team
            print(f"Team '{team_name}' added successfully.")
        except ValueError as e:
            print(f"Error: Invalid input type. Please enter numbers for counts. Details: {e}")

    
    def add_matchrecord(self, *args):
        
        if args:
            
            try:
               
                team_name, captain_name, num_batsmen, num_bowlers, num_allrounders, opponent_team, runs_scored, match_result, match_date = args
                new_record = MatchRecord(*args)
                self.match_records.append(new_record)
                print(f"Match record added successfully for {team_name} vs {opponent_team}.")
            except ValueError as e:
                 print(f"Error adding match record from parameters: Invalid data type or missing parameters. Details: {e}")
            except Exception as e:
                print(f"An unexpected error occurred: {e}")
        else:
          
            try:
                print("--- Enter Match Record Details ---")
                team_name = input("Enter your team's name: ")
                if team_name not in self.teams:
                    print(f"Error: Team '{team_name}' not found. Please add the team first.")
                    return
               
                team = self.teams[team_name]
                opponent_team = input("Enter opponent team's name: ")
                runs_scored = int(input("Enter runs scored by your team: "))
                match_result = input("Enter match result (e.g., Won, Lost, Draw): ")
                match_date = input("Enter match date (YYYY-MM-DD): ") # Simple string input for date

                new_record = MatchRecord(team.team_name, team.captain_name, team.num_batsmen, team.num_bowlers, team.num_allrounders,
                                         opponent_team, runs_scored, match_result, match_date)
                self.match_records.append(new_record)
                print(f"Match record added successfully for {team_name}.")
            except ValueError as e:
                print(f"Error: Invalid input type. Please ensure runs scored is a number. Details: {e}")
            except Exception as e:
                print(f"An unexpected error occurred: {e}")


    def update_matchrecord(self):
       
        team_name = input("Enter team name to update records for: ")
        found = False
        for record in self.match_records:
            if record.team_name == team_name:
                print(f"Current Record for match against {record.opponent_team} on {record.match_date}: Result: {record.match_result}")
                new_result = input("Enter new result: ")
                record.match_result = new_result
                print("Record updated.")
                found = True
        if not found:
            print(f"No match records found for team '{team_name}'.")

    def search_team(self):
       
        team_name = input("Enter team name to search: ")
        if team_name in self.teams:
            self.teams[team_name].display()
        else:
            print(f"Team '{team_name}' not found.")

    def display_match_record(self):
        
        if not self.match_records:
            print("No match records found.")
        for record in self.match_records:
            record.display()
            print("-" * 20)

    def add_player(self):
        
        try:
            player_id = input("Enter player ID: ")
            if player_id in self.players:
                print(f"Error: Player with ID '{player_id}' already exists.")
                return
            player_name = input("Enter player name: ")
            team_name = input("Enter the team name the player belongs to: ")
            if team_name not in self.teams:
                print(f"Error: Team '{team_name}' not found. Please add the team first.")
                return

            new_player = Player(player_id, player_name, team_name)
            self.players[player_id] = new_player
            print(f"Player '{player_name}' added successfully to '{team_name}'.")
        except Exception as e:
            print(f"An error occurred while adding player: {e}")

    def display_players(self):
        
        if not self.players:
            print("No players found.")
        for player in self.players.values():
            player.display()
            print("-" * 20)

    def menu(self):
       
        while True:
            print("\nTournament System Menu:")
            print("1. Add Team")
            print("2. Add Match Record (Manual Input)")
            print("3. Add Match Record (Using Function Overload Example Data)")
            print("4. Update Match Record")
            print("5. Search Team")
            print("6. Display All Match Records")
            print("7. Add Player")
            print("8. Display All Players")
            print("9. Exit")

            choice = input("Enter your choice: ")
            if choice == '1':
                self.add_team()
            elif choice == '2':
                self.add_matchrecord()
            elif choice == '3':
                
                if 'Royal Challengers Bangalore' not in self.teams:
                     print("Please add the 'Royal Challengers Bangalore' team first via Option 1.")
                else:
                    team_details = self.teams['Royal Challengers Bangalore']
                    self.add_matchrecord(team_details.team_name, team_details.captain_name, team_details.num_batsmen,
                                         team_details.num_bowlers, team_details.num_allrounders, 'Chennai Super kings', 180, 'Won', '2026-03-15')
            elif choice == '4':
                self.update_matchrecord()
            elif choice == '5':
                self.search_team()
            elif choice == '6':
                self.display_match_record()
            elif choice == '7':
                self.add_player()
            elif choice == '8':
                self.display_players()
            elif choice == '9':
                print("Exiting Tournament System.")
                break
            else:
                print("Invalid choice. Please try again.")

if __name__ == "__main__":
    system = TournamentSystem()
    
    system.teams['Royal Challengers Bangalore'] = Team('Royal Challengers Bangalore', 'Virat', 4, 3, 2)
    print("Pre-added 'Royal Challengers Bangalore' team for demonstration purposes.")
    system.menu()
```
